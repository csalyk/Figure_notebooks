## Figure 4.2: Lagrange points

```python
#Sun-Jupiter system
M1=1000.
M2=1.
R=10

#L1
r_L1=R*(M2/(3*M1))**(1./3.)
x_L1=R+r_L1
y_L1=0

#L2
x_L2=R-r_L1
y_L2=0

#L3
r_L3=R+(R*(5*M2/(12*M1)))
x_L3=-r_L3
y_L3=0

#L4
x_L4=R*np.cos(60*np.pi/180.)
y_L4=R*np.sin(60*np.pi/180.)

#L5
x_L5=R*np.cos(60*np.pi/180.)
y_L5=-R*np.sin(60*np.pi/180.)

fig=plt.figure(figsize=(5,5))
ax=fig.add_subplot(111,aspect=1)
ax.plot([0],[0],'ko',ms=14)
ax.plot([10],[0],'ko',ms=6)
ax.plot([x_L1],[y_L1],'rx',ms=6)
ax.plot([x_L2],[y_L2],'rx',ms=6)
ax.plot([x_L3],[y_L3],'rx',ms=6)
ax.plot([x_L4],[y_L4],'rx',ms=6)
ax.plot([x_L5],[y_L5],'rx',ms=6)

offset=1.05
ax.text(x_L1*offset,y_L1*offset,r'L$_2$',fontsize=14,fontweight='bold',color='red')
ax.text(x_L2*0.8,y_L2*offset,r'L$_1$',fontsize=14,fontweight='bold',color='red')
ax.text(x_L3/offset,y_L3*offset,r'L$_3$',fontsize=14,fontweight='bold',color='red')
ax.text(x_L4*offset,y_L4*offset,r'L$_4$',fontsize=14,fontweight='bold',color='red')
ax.text(x_L5*offset,y_L5/offset,r'L$_5$',fontsize=14,fontweight='bold',color='red')
ax.set_axis_off()

ax.text(-0.7,-2,'m$_1$',fontsize=14)
ax.text(9.5,-2,'m$_2$',fontsize=14)

ax.set_xlim(-12,12)
ax.set_ylim(-12,12)

fig.savefig(fig_dir+'Fig04x02.eps',bbox_inches='tight')
```
