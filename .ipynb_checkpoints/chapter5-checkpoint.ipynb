## Figure 5.1: Star/planet contrast

```python
planet_data=pd.read_csv(data_dir+'planet_data.csv')

lam=np.logspace(-2, 4, num=100)*1e-6
nu=c.value/lam
planet_temps=planet_data['Mean Temperature']
planet_names=planet_data['Name']

T=5780.
I_nu_sun=2*np.pi*h.value*nu**3./(c.value**2.)*(1.)/(np.exp(h.value*nu/(k_B.value*T))-1)
T=planet_temps[2]
I_nu_Earth=2*np.pi*h.value*nu**3./(c.value**2.)*(1.)/(np.exp(h.value*nu/(k_B.value*T))-1)
T=planet_temps[4]
I_nu_Jupiter=2*np.pi*h.value*nu**3./(c.value**2.)*(1.)/(np.exp(h.value*nu/(k_B.value*T))-1)
T=planet_temps[7]
I_nu_Neptune=2*np.pi*h.value*nu**3./(c.value**2.)*(1.)/(np.exp(h.value*nu/(k_B.value*T))-1)

albedo_Jupiter=0.503
albedo_Earth=0.306
albedo_Neptune=0.290

radii=planet_data['Diameter']/2.
R_sun=0.0046*au
R_Earth=radii[2]
R_Jupiter=radii[4]
R_Neptune=radii[7]

semimajoraxes=planet_data['Semi-major axis']
a_Earth=semimajoraxes[2]*au
a_Jupiter=semimajoraxes[4]*au
a_Neptune=semimajoraxes[7]*au

L_nu_sun=I_nu_sun*4*np.pi*R_sun**2.
L_nu_Earth=I_nu_Earth*4*np.pi*R_Earth**2. + I_nu_sun*(R_sun/a_Earth)**2.* np.pi*R_Earth**2.*albedo_Earth
L_nu_Jupiter=I_nu_Jupiter*4*np.pi*R_Jupiter**2.+ I_nu_sun*(R_sun/a_Jupiter)**2.* np.pi*R_Jupiter**2.*albedo_Jupiter
L_nu_Neptune=I_nu_Neptune*4*np.pi*R_Neptune**2.+ I_nu_sun*(R_sun/a_Neptune)**2.* np.pi*R_Neptune**2.*albedo_Neptune

fig=plt.figure(figsize=(10,6))
ax1=fig.add_subplot(111)
ax1.plot(lam*1e6, L_nu_sun,label="Sun",color='C0',linestyle='-',linewidth=4)
ax1.plot(lam*1e6, L_nu_Earth, label="Earth",color='C1',linestyle='--',linewidth=4)
ax1.plot(lam*1e6, L_nu_Jupiter, label="Jupiter",color='C2',linestyle='-.',linewidth=4)
ax1.plot(lam*1e6, L_nu_Neptune, label="Neptune",color='C3',linestyle=':',linewidth=4)

for tick in ax1.xaxis.get_major_ticks():
    tick.label1.set_fontsize(16)
    tick.label1.set_fontweight('medium')
for tick in ax1.yaxis.get_major_ticks():
    tick.label1.set_fontsize(16)
    tick.label1.set_fontweight('medium')

ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_ylim(1e-4,1e14)
ax1.set_xlim(4e-2,1e3)

ax1.set_ylabel(r'L$_\nu$ [W Hz$^{-1}$]',fontsize=20)
ax1.set_xlabel('Wavelength [$\mu$m]',fontsize=20)
ax1.legend(fontsize=14,handlelength=4)

fig.savefig(fig_dir+'Fig05x01.eps',bbox_inches='tight')
```

## Figure: Habitable zone

```python
def T_eq(L_star, a_planet):
    return (L_star*L_sun.value/(16*np.pi*sigma_sb.value*(a_planet*au.value)**2))**(0.25)

def a_habitable(L_star, T):
    return np.sqrt(L_star*L_sun.value/(16*np.pi*sigma_sb.value*T**4.))/au.value

def L_habitable(a_habitable,T):
    return 16*np.pi*sigma_sb.value*a_habitable**2*T**4.

out=pd.read_csv(data_dir+'nasa_exoplanets.csv',header=147)
alist=np.linspace(0,2,100)
lum250=L_habitable(alist*au.value,250.)/L_sun.value
lum300=L_habitable(alist*au.value,300.)/L_sun.value
lum350=L_habitable(alist*au.value,350.)/L_sun.value

loglum=out['st_lum']
lum=10**loglum
a=out['pl_orbsmax']
my_bool=(np.isfinite(loglum) & np.isfinite(a))

ss_lum=np.zeros(9)+1
ss_a=planet_data['Semi-major axis'][0:9]

fig=plt.figure(figsize=(8,5))
ax1=fig.add_subplot(111)
ax1.plot(alist,lum250,'--',color='C0')
ax1.plot(alist,lum300,color='C0')
ax1.plot(alist,lum350,'--',color='C0')
ax1.fill_between(alist,lum250,lum350,color='C0',alpha=0.1)
ax1.plot(a[my_bool],lum[my_bool],'.',color='C1',label='Exoplanets')
ax1.plot(ss_a, ss_lum,'*',color='C4',label='Solar system planets',ms=10)
ax1.legend(fontsize=16)
ax1.set_xlim(0,2.)
ax1.set_ylim(0.1,10.)
ax1.set_yscale('log')
ax1.set_yticks((0.1,1,10))
ax1.set_yticklabels(('0.1','1','10'))
for tick in ax1.xaxis.get_major_ticks():
    tick.label1.set_fontsize(16)
    tick.label1.set_fontweight('medium')
for tick in ax1.yaxis.get_major_ticks():
    tick.label1.set_fontsize(16)
    tick.label1.set_fontweight('medium')
ax1.text(1,5,'T$_\mathrm{eq}$=350 K',fontsize=14,color='C0')
ax1.text(1.3,3.2,'T$_\mathrm{eq}$=300 K',fontsize=14,color='C0')
ax1.text(1.5,2,'T$_\mathrm{eq}$=250 K',fontsize=14,color='C0')

ax1.set_xlabel('Planet semi-major axis [AU]',fontsize=18)
ax1.set_ylabel('Stellar luminosity [L$_\odot$]',fontsize=18)
fig.set_rasterized(True)
fig.savefig(fig_dir+'Fig05x19.eps',bbox_inches='tight')
```
