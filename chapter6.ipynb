## Figure 6.2: Determining distance to orbits

```python
circle_earth = Ellipse(np.array([0,0]),width=10, height=10,
                 angle=0,facecolor='None',edgecolor='k',lw=1)  
circle_int1 = Ellipse(np.array([0,0]),width=10*0.723, height=10*0.723,
                 angle=0,facecolor='None',edgecolor='k',lw=1)  
circle_int2 = Ellipse(np.array([0,0]),width=10*0.387, height=10*0.387,
                 angle=0,facecolor='None',edgecolor='k',lw=1)  
x1=np.arange(0,10)
tan1a=-5+0.93*(x1)
tan1b=-1*(x1)

x2=-1*np.arange(0,10)
tan2a=-5-2.33*(x2)
tan2b=0.43*(x2)

fig=plt.figure(figsize=(10,6))
ax1=fig.add_subplot(111,aspect=1)
ax1.add_artist(circle_earth)
ax1.add_artist(circle_int1)
ax1.add_artist(circle_int2)
ax1.plot(0,-5,'o',color='black',ms=10)
ax1.plot(0,0,'y*',ms=10)

ax1.plot(x1,tan1a,color='C0')
ax1.plot(x1,tan1b,color='C0')
ax1.plot(2.55,-2.55,'o',color='C0',ms=10)

ax1.plot(x2,tan2a,color='C1')
ax1.plot(x2,tan2b,color='C1')
ax1.plot(-1.8,-0.75,'o',color='C1',ms=10)

ax1.plot([0,0],[-5,10],color='yellow')
ax1.text(0.3,-3.2,r'$\theta_\mathrm{max}$',color='C0',fontsize=14)
ax1.text(-0.8,-2.5,r'$\theta_\mathrm{max}$',color='C1',fontsize=14)
ax1.text(2.8,-2.7,'V',color='C0')
ax1.text(-2.5,-0.7,'M',color='C1')
ax1.text(0,-5.6,'E')

ax1.set_xlim(-6,6)
ax1.set_ylim(-6,6)
ax1.set_axis_off()
fig.savefig(fig_dir+'Fig06x02.eps',bbox_inches='tight')
```
