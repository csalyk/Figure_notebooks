## Figure 3.3a: Components of velocity vector

```python
fig=plt.figure(figsize=(6,4))
ecc=0.5
a=3
ell =  Ellipse(np.array([0,0]),
               width=2*a, height=2*a*np.sqrt(1-ecc**2.),
               angle=0,facecolor='none',edgecolor='gray',lw=1)
fig=plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(1,2,1,  aspect=1)
ax1.add_artist(ell)
ax1.plot(np.array([0])-a*ecc,[0],'y*',markersize=14)
ax1.plot([-a*ecc,1.9],[0,2],color='gray',linestyle='--')
ax1.quiver([1.9], [2], [-2], [1.15], scale_units='xy',angles='xy',scale=1)
ax1.quiver([1.9], [2], [-1], [-0.6], scale_units='xy',angles='xy',scale=1)
ax1.quiver([1.9], [2], [-1.15], [1.7], scale_units='xy',angles='xy',scale=1)
vlength=np.sqrt(2**2.+1.15**2.)
ax1.text(0.3,0.5,'r',color='k',fontsize=12)
ax1.text(-0.1,3.2,r'$\vec{v}$',fontsize=15)
ax1.text(1.6,1.2,r'$\vec{v_r}$',fontsize=15)
ax1.text(1.5,3,r'$\vec{v_\theta}$',fontsize=15)
ax1.set_axis_off()
ax1.set_xlim(-3.2, 3.2)
ax1.set_ylim(-3.2, 3.8)
fig.savefig(fig_dir+'Fig03x03a.eps',bbox_inches='tight')
```

## Figure 3.3b: Kepler's 2nd law triangle

```python
ecc=0.5
a=3
ell =  Ellipse(np.array([0,0]),
               width=2*a, height=2*a*np.sqrt(1-ecc**2.),
               angle=0,facecolor='none',edgecolor='gray',lw=1)
fig=plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(1,2,1,  aspect=1)
ax1.add_artist(ell)
ax1.plot(np.array([0])-a*ecc,[0],'y*',markersize=14)
ax1.plot([-a*ecc,2.88],[0,0.7],color='black')
ax1.plot([-a*ecc,2.76],[0,1],color='black')
#ax1.text(3,0.8,r'$\Delta\, s$',color='blue',fontsize=12)
ax1.text(0.5,-0.1,'r',color='k',fontsize=12)
ax1.plot([2.88,2.76],[0.7,1],color='black',lw=3)
ax1.set_xlim(-3.2, 3.2)
ax1.set_ylim(-3.2, 3.2)
ax1.set_axis_off()

ax2 = fig.add_subplot(1,2,2)
ax2.plot([0,3],[0,0],color='black')
ax2.plot([0,2.97],[0,0.2],color='black')
ax2.text(3.1,0.1,r'r $\Delta\theta$',color='black',fontsize=14)
ax2.text(1.5,-0.1,'r',color='black',fontsize=14)
ax2.plot([3,2.97],[0,0.2],color='black',lw=3)
ax2.set_xlim(-0.1, 3.2)
ax2.set_ylim(-0.5, 0.5)
ax2.set_axis_off()
ax2.text(1.,0.02,r'$\Delta\theta$',fontsize=13)
fig.savefig(fig_dir+'Fig03x03b.eps',bbox_inches='tight')
```
