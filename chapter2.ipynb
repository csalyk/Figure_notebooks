## Figure 2.2: Blackbody curves

```python
lam_max_100=2898./100.
lam_max_1000=2898./1000.
lam_max_10000=2898./10000.

lam=np.logspace(-2, 5, num=100)*1e-6
nu=c.value/lam

T=100.
I_nu_100=2*np.pi*h.value*nu**3./(c.value**2.)*(1.)/(np.exp(h.value*nu/(k_B.value*T))-1)
T=1000.
I_nu_1000=2*np.pi*h.value*nu**3./(c.value**2.)*(1.)/(np.exp(h.value*nu/(k_B.value*T))-1)
T=10000.
I_nu_10000=2*np.pi*h.value*nu**3./(c.value**2.)*(1.)/(np.exp(h.value*nu/(k_B.value*T))-1)

fig=plt.figure(figsize=(10,6))
ax1=fig.add_subplot(111)
ax1.plot(lam*1e6, I_nu_10000,label="T=10,000 K",color='blue',linestyle='--',lw=4)
ax1.plot(lam*1e6, I_nu_1000, label="T=1,000 K",color='green',linestyle='-.',lw=4)
ax1.plot(lam*1e6, I_nu_100, label="T=100 K",color='red',linestyle='-',lw=4)

for tick in ax1.xaxis.get_major_ticks():
    tick.label1.set_fontsize(16)
    tick.label1.set_fontweight('medium')
for tick in ax1.yaxis.get_major_ticks():
    tick.label1.set_fontsize(16)
    tick.label1.set_fontweight('medium')

ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlim(1e-2,5e4)
ax1.set_ylim(1e-20,1e-6)
ax1.set_ylabel(r'I$_\nu$ [W m$^{-2}$ Hz$^{-1}$]',fontsize=18)
ax1.set_xlabel('Wavelength [$\mu$m]',fontsize=18)
ax1.legend(fontsize=14,handlelength=4)

fig.savefig(fig_dir+'Fig02x02.eps',bbox_inches='tight')
```
