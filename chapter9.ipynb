## Figure 9.19: Chondrite abundances

```python
out=pd.read_csv(data_dir+'abundances.txt',dtype={'Num':"float64",'Elem':str,'Phot':"float64",
                                        'Phot_Err':"float64",'Met':"float64",
                                        'Met_Err':"float64"},header=0,skipinitialspace=True)
fin=np.abs(out['Phot'])>0.01
phot=np.array(out['Phot'][fin])
met=np.array(out['Met'][fin])
elem=np.array(out['Elem'][fin])

out2=pd.read_csv(data_dir+'earth_abundances.txt',dtype={'Elem':str,'Abun':"float64",'Err':"float64",
                                        'Factor':"float64"},header=0)
earth0=np.array(out2['Abun'])*np.array(out2['Factor'])
earth_elem=np.array(out2['Elem'])
print(out[out['Elem']=='Mo']['Phot'])  #Log of photospheric abundance
print(out2[out2['Elem']=='Mo']['Abun']*out2[out2['Elem']=='Mo']['Factor']) #Earth abundance
normfactor=(10**1.88)/0.000002  #Mo in photosphere divided by Si on Earth
print(normfactor)
earth=[]
#Loop through phot elements.  Find Earth value
for myelem in elem:
    mybool=(earth_elem==myelem)
    myearth=np.log10(out2['Abun'][mybool]*out2['Factor'][mybool]*normfactor)
    if(np.sum(mybool)==0): 
        earth.append(0)
    if(np.sum(mybool)==1): 
        earth.append(myearth.item())
earth=np.array(earth)
fin2=((np.abs(phot)>0.01) & (earth!=0))
```

```python
fig=plt.figure(figsize=(6,10))
ax1=fig.add_subplot(211)
ax1.plot(phot, met,'C0o')
ax1.plot([-3,11],[-3,11],'--')
for tick in ax1.xaxis.get_major_ticks():
    tick.label1.set_fontsize(16)
    tick.label1.set_fontweight('medium')
for tick in ax1.yaxis.get_major_ticks():
    tick.label1.set_fontsize(16)
    tick.label1.set_fontweight('medium')
for i,mylabel in enumerate(elem):
    if(np.abs(phot[i]-met[i])>0.3):
        ax1.text(phot[i],met[i],mylabel,fontsize=12)
ax1.set_ylabel('Log CI Chondrite Abundance',fontsize=14)
ax1.set_xlim(-3,13)
ax1.set_ylim(-2.5,10)

ax2=fig.add_subplot(212)
ax2.plot(phot[fin2], earth[fin2],'C0o')
for i,mylabel in enumerate(elem):
    if((np.abs(phot[i]-earth[i])>0.3) &  (earth[i]>-2) & fin2[i]==True):
        ax2.text(phot[i],earth[i],mylabel,fontsize=12)
ax2.plot([-3,11],[-3,11],'--')
ax2.set_xlim(-3,13)
ax2.set_ylim(-2.5,10)
ax2.set_xlabel('Log Solar Abundance',fontsize=14)
for tick in ax2.xaxis.get_major_ticks():
    tick.label1.set_fontsize(16)
    tick.label1.set_fontweight('medium')
for tick in ax2.yaxis.get_major_ticks():
    tick.label1.set_fontsize(16)
    tick.label1.set_fontweight('medium')
ax2.set_ylabel('Log Bulk Earth Abundance',fontsize=14)

fig.savefig(fig_dir+'Fig09x19.eps',bbox_inches='tight')
```
