## Figure 1.1: Number of planets vs. time

```python
years=np.array([1700,1781,1801,1802, 1804, 1807, 1840, 1846, 1860,1915,1930,2006])
number=np.array([6,   7,   8,   9,   10,   11,    7,    8,     9,   8,   9,    8])

notes=([' ','Uranus','Ceres','Pallas','Juno', 'Vesta', 
        'Demotion of asteroids','Neptune', 'Vulcan',  
        'Removal of Vulcan','Pluto','Demotion of Pluto'])

fig = plt.figure(figsize=(10, 5))
ax=fig.add_subplot(111)
ax.plot(years, number, 'o',markersize=10)
ax.plot(years, number,linewidth=3)
ax.set_ylim(5,15)
ax.set_xlim(1695,2015)
ax.set_xlabel('Year', fontweight='medium', fontsize=20)
ax.set_ylabel('Number of known planets', fontweight='medium', fontsize=20)
for i,year in enumerate(years[0:12]):
    if(notes[i]=='Demotion of asteroids'):
        ax.text(year-20, number[i]-0.7,'Demotion of', fontsize=14)
        ax.text(year-20, number[i]-1.3,'asteroids', fontsize=14)

    if(notes[i]=='Demotion of Pluto'):
        ax.text(year-45, number[i]-0.6, 'Demotion of', fontsize=14)
        ax.text(year-45, number[i]-1.2, 'Pluto', fontsize=14)

    if(notes[i]=='Removal of Vulcan'):
        ax.text(year-15, number[i]-0.6, 'Removal of', fontsize=14)
        ax.text(year-15, number[i]-1.2, 'Vulcan', fontsize=14)
        
    if((notes[i]!='Removal of Vulcan') & (notes[i]!='Demotion of Pluto') & (notes[i]!='Demotion of asteroids')):
        ax.text(year+1, number[i]+0.1, notes[i], fontsize=14)

ax.tick_params(axis='y',  width=2, labelsize='x-large')
fontsize=14
for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('medium')
for tick in ax.yaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('medium')
fig.savefig(fig_dir+'Fig01x01.eps',bbox_inches='tight')
```

## Figure 1.4: Definition of ellipse and semi-major axis

```python
from matplotlib.patches import Ellipse

a=3.
b=2.
e = Ellipse(np.array([0,0]),
                width=a*2, height=b*2,
                angle=0)
ecc=np.sqrt(1-(b/a)**2)

fig=plt.figure(figsize=(5,5))
ax=fig.add_subplot(111)
ax.add_artist(e)
e.set_clip_box(ax.bbox)
e.set_alpha(0.4)
ax.set_axis_off()
ax.plot([-1*a*(ecc)],[0],'k*',markersize=12)
ax.set_xlim(-3.2, 3.2)
ax.set_ylim(-3.2, 3.2)
ax.plot([0,3],[0,0],'k')
ax.plot([0,0],[0,2],'k')
ax.plot([3],[0],'ko')
ax.text(3.1,0,'Aphelion',fontsize=10)
ax.plot([-3],[0],'ko')
ax.text(-3.6,0.1,'Perihelion',fontsize=10)
ax.text(0.2,0.2,'Semi-major axis (a)',fontsize=12)
ax.text(-0.4,1.7,'Semi-minor axis (b)',rotation=90,fontsize=12)
fig.set_rasterized(True)
fig.savefig(fig_dir+'Fig01x04.eps',bbox_inches='tight')
```

## Figure 1.5: Bode's Law

```python
#Actual data for planets
planet_data=pd.read_csv(data_dir+'planet_data.csv')
a_planets=np.array(planet_data['Semi-major axis'][:-1])
name_planets=np.array(planet_data['Name'][:-1])
a_planets=np.insert(a_planets,4,2.5, axis=None)
name_planets=np.insert(name_planets,4,'Asteroids',axis=None)

idx_planets=np.arange(0,10.,1.)-1

#Bode's law
n_Bode=np.array([-1e10,0,1,2,3,4,5,6,7,8])
a_Bode=0.4+0.3*2**n_Bode

fig=plt.figure(1,figsize=(13,5))
ax1=fig.add_subplot(111)
ax1.plot(idx_planets, a_Bode,label=r"Bode's law : $0.4+0.3 \times 2^n$")
ax1.plot(idx_planets, a_planets, 'ro', label="Actual data")
ax1.set_xlim(-1.9,9)
ax1.set_xlabel('Index (n)',fontsize=18)
ax1.set_ylabel('Semi-major axis [AU]',fontsize=18)
x_formatter = matplotlib.ticker.ScalarFormatter(useOffset=False)
ax1.xaxis.set_major_formatter(x_formatter)
for tick in ax1.xaxis.get_major_ticks():
    tick.label1.set_fontsize(16)
    tick.label1.set_fontweight('medium')
for tick in ax1.yaxis.get_major_ticks():
    tick.label1.set_fontsize(16)
    tick.label1.set_fontweight('medium')    

ax1.legend(loc='upper left',fontsize=20)
for i, (x, y) in enumerate(zip(idx_planets, a_planets)):
    ax1.annotate(name_planets[i],xy=(x-0.2,y+2), xycoords='data',fontsize=12)
fig.savefig(fig_dir+'Fig01x05.eps',bbox_inches='tight')
```

## Figure 1.6: Different eccentricities

```python
from matplotlib.patches import Ellipse

ell_circle = Ellipse(np.array([0,0]),
                width=6, height=6,
                angle=0,facecolor='none',edgecolor='b',lw=4)

e_Earth = 0.017
ell_Earth = Ellipse(np.array([0,0]),
                width=6, height=6*np.sqrt(1-e_Earth**2.),
                angle=0,facecolor='none',edgecolor='b',lw=4)

e_Pluto = 0.244 #May want to check this
ell_Pluto =  Ellipse(np.array([0,0]),
                width=6, height=6*np.sqrt(1-e_Pluto**2.),
                angle=0,facecolor='none',edgecolor='b',lw=4)

e_Halley = 0.967
ell_Halley =  Ellipse(np.array([0,0]),
                width=6, height=6*np.sqrt(1-e_Halley**2.),
                angle=0,facecolor='none',edgecolor='b',lw=4)

fig=plt.figure(figsize=(14,10))

#Circle
ax1 = fig.add_subplot(1,4,1,  aspect=1)
ax1.add_artist(ell_circle)
ell_circle.set_alpha(0.8)
ax1.plot(np.array([0]),[0],'k.',markersize=5)
ax1.plot([0],[0],'k*',markersize=14)
ax1.set_xlim(-3.2, 3.2)
ax1.set_ylim(-3.2, 3.2)
ax1.text(-2.4,1,'Circle (e=0)',fontsize=16)
ax1.set_axis_off()

#Earth
ax2 = fig.add_subplot(1,4,2,  aspect=1)
ax2.add_artist(ell_Earth)
ell_Earth.set_alpha(0.8)
ax2.plot(np.array([0]),[0],'k.',markersize=5)
ax2.plot(np.array([0])+3*e_Earth,[0],'k*',markersize=14)
ax2.set_xlim(-3.2, 3.2)
ax2.set_ylim(-3.2, 3.2)
ax2.text(-2.4,1,'Earth (e=0.017)',fontsize=16)
ax2.set_axis_off()

#Pluto
ax3 = fig.add_subplot(1,4,3,  aspect=1)
ax3.add_artist(ell_Pluto)
ell_Pluto.set_alpha(0.8)
ax3.plot(np.array([0]),[0],'k.',markersize=5)
ax3.plot(np.array([0])+3*e_Pluto,[0],'k*',markersize=14)
ax3.set_xlim(-3.2, 3.2)
ax3.set_ylim(-3.2, 3.2)
ax3.text(-2.4,1.,'Pluto (e=0.244)',fontsize=16)
ax3.set_axis_off()

#Halley
ax4 = fig.add_subplot(1,4,4,  aspect=1)
ax4.add_artist(ell_Halley)
ell_Halley.set_alpha(0.8)
ax4.plot(np.array([0]),[0],'k.',markersize=5)
ax4.plot(np.array([0])+3*e_Halley,[0],'k*',markersize=14)
ax4.set_xlim(-3.2, 3.2)
ax4.set_ylim(-3.2, 3.2)
ax4.text(-2.8,1,"Halley's comet (e=0.967)",fontsize=16)
ax4.set_axis_off()
fig.set_rasterized(True)
fig.savefig(fig_dir+'Fig01x06.eps',bbox_inches='tight')
```

## Figure 1.7: Inclinations with respect to the invariable plane.

```python
planet_data=pd.read_csv(data_dir+'planet_data.csv')
name_planets=np.array(planet_data['Name'][:-1])

inc_ip=np.array([6.3,2.2,1.6,1.7,0.3,0.9,1.0,0.7,15.6])
fig=plt.figure(figsize=(10,5))
ax1=fig.add_subplot(111,aspect=1)

for i,myinc in enumerate(inc_ip):
    y1=10*np.tan(myinc*np.pi/180.)
    if(name_planets[i]=='Mercury'):
        ax1.plot([-10,10],[-y1,y1],'-.',label=name_planets[i]+r' (6.3$^\circ$)',lw=4)
    if(name_planets[i]=='Pluto'):
        ax1.plot([-10,10],[-y1,y1],'--',label=name_planets[i]+r' (15.6$^\circ$)',lw=4)
    if((name_planets[i]!='Pluto') and (name_planets[i] !='Mercury')):
        ax1.plot([-10,10],[-y1,y1])

ax1.legend(loc='upper left',fontsize=14)
ax1.set_axis_off()

fig.savefig(fig_dir+'Fig01x07.eps',bbox_inches='tight')
```
